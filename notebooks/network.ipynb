# Network data structure

`compas.datastructures`

The network data structure is an edge-oriented graph. It contains nodes (or vertices) connected by links (or edges).

* A network can consist of only vertices.
* A network has no faces.
* A network can have crossing edges.
* ...

```python
import compas

from compas.datastructures import Network
from compas.plotters import NetworkPlotter

network = Network.from_obj(compas.get('grid_irregular.obj'))

plotter = NetworkPlotter(network, figsize=(8, 6))
plotter.draw_vertices()
plotter.draw_edges()
plotter.show()
```

**Topology**

```python
for key in network.vertices():
    print(key, network.vertex_neighbours(key))
```

**Visualisation**

```python
plotter = NetworkPlotter(network, figsize=(8, 6))

plotter.draw_vertices(
    facecolor={key: '#ff0000' for key in network.leaves()},
    text={key: network.vertex_degree(key) for key in network.vertices()},
    radius=0.25
)

plotter.draw_edges()
plotter.show()
```
